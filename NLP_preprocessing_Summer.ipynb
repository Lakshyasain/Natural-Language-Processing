```python
import re
def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'',text)
```

Combining all together, '<.*?>' matches any HTML tag (including self-closing tags) and matches as few characters as possible between the opening and closing angle brackets.

The sub method of the regex object (pattern) is used to perform substitution. It replaces all occurrences of the regex pattern in the input text with the replacement string '', effectively removing the HTML tags.

```python
df['review'] = df['review'].apply(remove_html_tags)
```

```python
def remove_url(text):
  pattern = re.compile(r'https?://\S+//|www.\.\S+')
  return pattern.sub(r'',text)
```

```python
df['review'] = df['review'].apply(remove_url)
```

```python
import string,time
exclude = string.punctuation
exclude
```

```python
def remove_punc(text):
  for char in exclude:
    text = text.replace(char,'')
  return text
```

```python
df['review'] = df['review'].apply(remove_punc)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = stopwords.words('english')
stop_words
```

```python
def remove_stopwords(text):
  new_text = []

  for word in text.split():
    if word in stop_words:
      new_text.append('')
    else:
      new_text.append(word)
  x = new_text[:]
  new_text.clear()
  return " ".join(x)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
def stem_word(text):
  return " ".join([ps.stem(word) for word in text.split()])
```
